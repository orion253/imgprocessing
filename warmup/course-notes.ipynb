```python
W = imageObj.width
H = imageObj.height
imdata = np.zeros((H,W,3), dtype=np.uint8)
imdata[:,:,0] = np.reshape(imageObj.getdata(band=0), (H,W))
imdata[:,:,1] = np.reshape(imageObj.getdata(band=1), (H,W))
imdata[:,:,2] = np.reshape(imageObj.getdata(band=2), (H,W))
```
